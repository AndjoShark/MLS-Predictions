```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('modelingdf.csv')
df = df.drop('Unnamed: 0', axis = 1)
```

```python
df.head()
```

```python
heatmap = df.drop(['name', 'club', 'position'], axis = 1)
corr = heatmap.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
plt.pyplot.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='RdBu_r', linewidth = .5, mask = mask)
plt.pyplot.show()
```

```python
sns.barplot(x = 'season', y='next_year_compensation', data=df)
plt.pyplot.xticks(rotation=70)
plt.pyplot.show()
```

```python
sns.swarmplot(x='season', y = 'next_year_compensation', data=df)
plt.pyplot.xticks(rotation=70)
plt.pyplot.show()
```

```python
sns.boxplot(x='season', y='next_year_compensation', data = df)
plt.pyplot.xticks(rotation=70)
plt.pyplot.show()
```

```python
sns.boxplot(x='season', y='next_year_compensation', data = df)
plt.pyplot.xticks(rotation=70)
plt.pyplot.ylim(0, 1000000)
plt.pyplot.show()
```
