```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('df.csv')

#Dropping $ and , in total compensation and renaming to next_year_earnings
df['total_compensation'] = df['total_compensation'].str.replace('$' , '')
df['total_compensation'] = df['total_compensation'].str.replace(',' , '')
df['total_compensation'] = df['total_compensation'].astype(int)
df.rename(columns={'total_compensation': 'next_year_compensation'}, inplace=True)
```

```python
#creating name to use to get current year salary
df['name'] = df['first_name_cleaned'] + ' ' + df['last_name_cleaned']

#dropping repetitive columns made from scraping and merging data
dropped_columns = ['Unnamed: 0', 'Club', 'previous_year', 'url', 'urls', 'Year',
                  'club_grouped', 'first_name_cleaned', 'last_name_cleaned', 'base_salary'
                  , 'first_name', 'last_name']
df = df.drop(dropped_columns, axis = 1)
df.head().T
```

```python
# creating current compensation
from tqdm import tqdm

df['current_compensation'] = 0

for i,row in tqdm(df.iterrows()):
    if i < 10000:
        row_name = row['name']
        current_year = row['season'] - 1
        next_year = row['season']
        row_name_previous_year = df[(df['name'] == row_name) & (df['season'] == current_year)]
        if len(row_name_previous_year) < 1:
            df.loc[i, 'current_compensation'] = -10
        else:
            df.loc[i, 'current_compensation'] = row_name_previous_year['next_year_compensation'].values[0]
```

```python
df.head().T
```

```python
df['season'] = df['season'] - 1
df = df[df['current_compensation'] > -1]
df.shape
```

```python
df.to_csv('modelingdf.csv')
```
